```python
import requests
import bs4
import urllib.parse
import IPython.display
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from IPython.display import Image, display


section_dict = {
    '최신뉴스':'n0100',
    '정치':'n0200',
    '경제':'n0300',
    '사회':'n0400',    
    '세계':'n0500',
    'IT/과학':'n0600',
    }


def print_news(section_name):
    # IT/과학 뉴스
    print('print_news(\'{}\')'.format(section_name))
    section = section_dict[section_name]
    # req_param = section_name
    #
    if section == "n0100":
        url = "https://news.nate.com/recent?mid={}".format(section)
        print(f"=====> {url} {section_name} 뉴스 <=====")
        
    else :
        url = "https://news.nate.com/section?mid={}".format(section)
        print(f"=====> {url} {section_name} 뉴스 <=====")
        
    # 요청 헤더 설정 : 브라우저 정보
    req_header = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36"
    }

    # requests 의 get() 함수 호출하기
    res = requests.get(url, headers=req_header)
    # res.encoding = "utf-8"
    
    # 응답(response)이 OK 이면
    if res.ok:
        html = res.text
        soup = BeautifulSoup(html, "html.parser")
        div_tag_list = soup.select("div.mlt01")
        for div_tag in div_tag_list:
            a_tag = div_tag.find("a")
            link = a_tag["href"]

            h2_tag = div_tag.select_one("h2.tit")
            title = h2_tag.text.strip()
            
            
            img_tag = div_tag.select_one("img")
            img_url = img_tag["src"]
            base_url = "https://..."
            full_url = urljoin(base_url, img_url)
            link_url = urljoin(base_url, link)
            if img_url is None:
                print(link_url, title)
            else:
                img = Image(full_url)
                display(img)
                print(link_url, title)
            print()
                
    else:
        # 응답(response)이 Error 이면 status code 출력
        print(f"Error code = {res.status_code}")
```

```python
print_news('경제')
```
